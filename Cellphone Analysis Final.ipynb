```python
import matplotlib.pyplot as plt
import seaborn as sns # import the libraries for visuals
```

```python
sns.heatmap(dataset.corr(), annot = True, cmap = "rainbow")
```

```python
sns.pairplot(dataset)
```

```python
colorsforx = ["Red", "Green", "Black", "Purple", "Yellow", "Red", "Green", "Black", "Purple", "Yellow"]
colorchanger = 0
```

```python
for i in dataset.columns[:-1]:
    sns.scatterplot(data=dataset, x=i, y="Price", color=colorsforx[colorchanger])
    plt.show()
    colorchanger += 1

    # Reset colorchanger if it exceeds the length of colorsforx
    if colorchanger >= len(colorsforx):
        colorchanger = 0
```
